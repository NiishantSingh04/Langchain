### Streaming and Batch

### Streaming

Most models can stream their output content while it is being generated. By displaying output progessively, streaming significantly improves users experience, particularly for longer responses. Calling stream() returns an iterator that yeilds output chunks as they are produced. You csn udr s loop to process esch chunk in real-time:

```python
modelG.invoke("Write a 400 words paragraph on Philosophy")
```

```python
for chunk in modelG.stream("Write about the world's greatest Philosophers."):
    print(chunk.text, end="|", flush = True)
```

### Batch

Batching a collection of independent requests to a model can significantly improves performance and reduce the costs, as the processing can be done in parallel:

```python
responses = modelG.batch([
    "Why do parrots have colourful feathers?",
    "Why Owls do not make any noise while Flying?",
    "What is Quantum Computing?"
])
for response in responses:
    print(response)
```

```python
responses = modelG.batch([
    "Why do parrots have colourful feathers?",
    "Why Owls do not make any noise while Flying?",
   "What is Quantum Computing?"
],
config={
    'max_concurrancy': 5, #Limit to a 5 parallel calls
})
for response in responses:
    print(response)
```
