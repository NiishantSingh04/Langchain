### MESSAGES

Messages are the fundamental unit of context for models in LangChain. They represent the input and output of models, carrying both the content and the metadata needed to represent the state of a conversation when interacting wiht an LLM. Messages are objects that contains:

    •Role - Identifies the messages type (e.g. system, user)

    •Content - represent the actual content of the message (like text, images, audio, documents, etc)

    •MetaData - Optional fields such as response information, messages IDs, and token usages

LangChain provides a standard messages type that works across all model providers, ensuring consistent behavior regardless of the model being called.

```python
import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

model = init_chat_model("groq:qwen/qwen3-32b")
```

```python
model.invoke("Please tell me what is artifical intelligence")
```

### MESSAGE PROMPT

Alternatively, you can pass in a list of messages to the model by providing a list of message objects.

### Messages Types - 
    •System Messages - Tells the model how to behave and provide context for interaction.
    •Human Messages - Represents user input and interaction with the model.
    •AI Messages - Responses generated by the model, including text content, tool calls, and metadata.
    •Tool Messgaes - Represents the outputs of tool calls.



### SYSTEM MESSAGES

A System Message represent an inital set of instructions that primes the model's behavior. You can use system messgae to set the tone, define the model's role, and establish guidelines for responses

### HUMAN MESSAGES

A Human Messages represent the user input and interactions. They can contain text, images, audio, files, and any other amounr of multimodal content

### AI MESSAGES

An AI Message represents the output of a modal invocation. They can include multimodal data, tool calls, and provider-specific metadata that you can later access.

### TOOL MESSAGES

For modals that support tool calling, AI Messages can contains tool calls. Tool messages are used to pass the result of a single tool executin back to model.

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages=[
    SystemMessage("You are a poetry expert."),
    HumanMessage("Tell me about Love.")
]

response = model.invoke(messages)
response.content
```

```python
system_msg = SystemMessage("You are a helpful coding assistant.")

messages = [
    system_msg,
    HumanMessage("How do I create a rest API?")
]

response = model.invoke(messages)
print(response.content)
```

```python
# Detailed Information to LLM through System Message

from langchain.messages import SystemMessage, HumanMessage, AIMessage

system_msg = SystemMessage("""You are a senior Python developer with the expertise in web frameworks.
Always provides code examples and explain your reasoning.
Be concise but thorough in your explanations.
""")

messages = [
    system_msg,
    HumanMessage("How do I create a RET API?")
]
response = model.invoke(messages)
print(response.content)
```

```python
# Message MetaData

human_msg = HumanMessage(
    content = "Hello!",
    name = "Alice",   #Optional: Identify different user
    id = "msg_123",   #Optional: Unique identifier for tracking
)
```

```python
response = model.invoke([
    human_msg
])
response
```

```python
from langchain.messages import AIMessage, SystemMessage, HumanMessage

#Create an AI message munually (e.g., for conversation history)
ai_msg = AIMessage("I'd be happy to help you with that question.")

#Add to conversation history 
messages = [
    SystemMessage("You are a ahelpful Assistant."),
    HumanMessage("Can you help me?"),
    ai_msg,  #Insert as if it came from model.
    HumanMessage("Great! What is 2+2?")
]

response = model.invoke(messages)
print(response.content)
```

```python
response.usage_metadata
```

```python
from langchain.messages import AIMessage, ToolMessage, HumanMessage

#After a model makes a tool call
# Here, we demonstrate mamually creating the message for brevity

ai_message = AIMessage(
    content = [],
    tool_calls=[{
        "name":"getweather",
        "args": {"location": "San Francisco"},
        "id": "call_123"
    }]
)

#Execute Tool and creare result messafe
weather_result = "Sunny, 72˚F"
tool_message = ToolMessage(
    content = weather_result,
    tool_call_id = "call_123" #Must match the call ID
)

#Continue conversation
messages = [
    HumanMessage("What's the weather in San Francisco?"),
    ai_message,  #Model's tool call
    tool_message,
]

response = model.invoke(messages)  #Model process result
print(response)
```
